### Reading IMM non-commercial data from CFTC

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
%matplotlib inline 
```

AUD_CFTC is the file of the transaction data from the CFTC. CFTC reports IMM non-commercial data every Tuesday.

```python
aud = pd.read_excel('AUD_CFTC.xlsx', index_col=0)
aud.head()
```

```python
df = aud.sort_index(ascending=True)
df.head()
```

```python
fig = plt.figure()
plt.plot(df)
plt.show()
```

```python
currency = pd.read_excel('AUD_USD_Historical_Data.xlsx', index_col=0)
audusd = currency.sort_index(ascending=True)
audusd.head()
```

```python
start=datetime.datetime(2012,1,1)
end = datetime.datetime(2018,3,31)
```

```python
aud=web.DataReader('DEXUSAL','fred',start,end)
aud.tail()
```

```python
fig, ax1 = plt.subplots(figsize=(8,4))

# fig, ax = plt.subplots(len(channels), sharex=True, figsize=(50,100))

color = 'black'
ax1.set_xlabel('Date')
ax1.set_ylabel('Price',color=color)
ax1.plot(aud['DEXUSAL'],color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'blue'
ax2.set_ylabel('Change',color=color)
ax2.plot(df['Dealer_Positions_Long_All'],color=color)
ax2.plot(df['Asset_Mgr_Positions_Long_All'],color='r')

ax1.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```

```python
gap_dealer = df['Dealer_Positions_Long_All']-df['Dealer_Positions_Short_All']
```

```python
fig,ax1 = plt.subplots(figsize=(10,4))
ax1.plot(df['Dealer_Positions_Long_All'],color='r')
ax1.plot(df['Dealer_Positions_Short_All'],color='g')
ax1.plot(gap_dealer,color='k')
plt.show()
```

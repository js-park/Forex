### Reading IMM non-commercial data from CFTC

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
aud = pd.read_excel('AUD_CFTC.xlsx', index_col=0)
aud.head()
```

```python
df = aud.sort_index(ascending=True)
df.head()
```

```python
fig = plt.figure()
plt.plot(df)
plt.show()
```

```python
fig = plt.figure()
plt.plot(df[['Dealer_Positions_Long_All','Asset_Mgr_Positions_Long_All']])
plt.show()
```

```python
import seaborn as sns
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```
